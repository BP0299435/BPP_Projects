```python
import pandas as pd
champ_data = pd.read_csv('export_lol_champs.csv')
lol_champions = pd.read_csv('lol_champs.csv')

# Display the first rows of each dataset
print("Champion Data (export_lol_champs.csv)")
print(champ_data.head())

print("\nChampion List (lol_champions.csv)")
print(lol_champions.head())
```

```python
# Display information about each dataset
print("Champion Data Info (export_lol_champs.csv):")
print(champ_data.info())
print("\nChampion Data Description (export_lol_champs.csv):")
print(champ_data.describe())

print("\nChampion Details Info (lol_champions.csv):")
print(lol_champions.info())
print("\nChampion Details Description (lol_champions.csv):")
print(lol_champions.describe())

# Check for missing values
print("\nMissing Values in Champion Data:")
print(champ_data.isnull().sum())

print("\nMissing Values in Champion Details:")
print(lol_champions.isnull().sum())

# Check for duplicate rows
print("\nDuplicate Rows in Champion Data:")
print(champ_data.duplicated().sum())

print("\nDuplicate Rows in Champion Details:")
print(lol_champions.duplicated().sum())
```

```python
# Drop redundant column
champ_data.drop(columns=["Unnamed: 0"], inplace=True)
lol_champions.drop(columns=["Unnamed: 0"], inplace=True)

# Convert 'AS+' to numeric, handling errors
champ_data['AS+'] = pd.to_numeric(champ_data['AS+'], errors='coerce')
lol_champions['AS+'] = pd.to_numeric(lol_champions['AS+'], errors='coerce')

# Handle missing 'Style' values
lol_champions['Style'].fillna("Unknown", inplace=True)

# Identify and clean numeric-looking columns with potential non-numeric entries
columns_to_check = ['Popularity', 'Winrate', 'Gold', 'DamageDealt']
for col in columns_to_check:
    lol_champions[col] = pd.to_numeric(lol_champions[col].str.replace('%', '').str.replace(',', ''), errors='coerce')

# Confirm changes
print("Champion Data (export_lol_champs.csv) after cleaning:")
print(champ_data.head())

print("\nChampion Details (lol_champions.csv) after cleaning:")
print(lol_champions.head())
```

```python
# Merge datasets on the 'Champions' column
merged_data = pd.merge(champ_data, lol_champions, on="Champions", how="inner")

# Display basic information about the merged dataset
print("Merged Data Info:")
print(merged_data.info())

# Display the first few rows to ensure the merge worked
print("\nMerged Data Preview:")
print(merged_data.head())
```

```python
# Dropping duplicate columns
merged_data.drop(columns=['HP_y', 'HP+_y', 'HP5_y', 'HP5+_y', 'MP_y', 'MP+_y',
                          'MP5_y', 'MP5+_y', 'AD_y', 'AD+_y', 'AS_y', 'AS+_y',
                          'AR_y', 'AR+_y', 'MR_y', 'MR+_y', 'MS_y', 'Range_y'],
                 inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Filter numeric columns
numeric_columns = merged_data.select_dtypes(include=['float64', 'int64'])

# Compute the correlation
correlation_matrix = numeric_columns.corr()

# Filter correlations greater than 0.6 or less than -0.6
filtered_corr = correlation_matrix[(correlation_matrix > 0.6) | (correlation_matrix < -0.6)]

# Plot the heatmap with rounded values
plt.figure(figsize=(15, 10))  # Increase the size for better readability
sns.heatmap(filtered_corr, annot=True, fmt=".1f", cmap="coolwarm", cbar=True, mask=filtered_corr.isnull())
plt.title("Filtered Correlation Heatmap (|correlation| > 0.6)")
plt.show()
```

```python
import numpy as np

# Convert numeric columns only
numeric_data = merged_data.select_dtypes(include=['float64', 'int64'])

# Recalculating the correlations with numeric data
updated_corr_matrix = numeric_data.corr()

# Filtering correlations above a certain threshold (|correlation| > 0.6)
threshold = 0.6
filtered_corr = updated_corr_matrix[
    (updated_corr_matrix >= threshold) | (updated_corr_matrix <= -threshold)
].copy()

# Removing self-correlations for better readability
np.fill_diagonal(filtered_corr.values, np.nan)

# Plotting the updated heatmap with cleaned formatting
plt.figure(figsize=(12, 8))
sns.heatmap(
    filtered_corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    cbar=True,
    linewidths=0.5,
    linecolor='black'
)
plt.title("Filtered Correlation Heatmap (|correlation| > 0.6)")
plt.tight_layout()
plt.show()
```

```python
# Enlarged heatmap with adjusted axis labels
plt.figure(figsize=(16, 10))  # Increase the figure size for better readability
sns.heatmap(filtered_corr, annot=True, fmt=".1f", cmap="coolwarm", square=True, cbar_kws={'shrink': 0.8})
plt.title("Filtered Correlation Heatmap (|correlation| > 0.6)", fontsize=16)
plt.xticks(rotation=45, fontsize=10, ha='right')  # Rotate and align the x-axis labels
plt.yticks(fontsize=10)  # Adjust font size for y-axis labels
plt.show()
```

```python
# Adding the classification column
merged_data['Preference'] = 'Neutral'

# Preferred champions
merged_data.loc[(merged_data['Popularity'] > merged_data['Popularity'].quantile(0.9)) &
                (merged_data['Winrate'] > 52), 'Preference'] = 'Preferred'

# Avoided champions
merged_data.loc[(merged_data['Popularity'] < merged_data['Popularity'].quantile(0.1)) &
                (merged_data['Winrate'] < 50), 'Preference'] = 'Avoided'
```

```python
merged_data['normalized_winrate'] = (merged_data['Winrate'] - merged_data['Winrate'].min()) / (merged_data['Winrate'].max() - merged_data['Winrate'].min())
merged_data['normalized_popularity'] = (merged_data['Popularity'] - merged_data['Popularity'].min()) / (merged_data['Popularity'].max() - merged_data['Popularity'].min())
merged_data['normalized_kda'] = (merged_data['KDA_kills'] + merged_data['KDA_assists']) / (merged_data['KDA_deaths'] + 1)  # Avoid division by zero
merged_data['normalized_kda'] = (merged_data['normalized_kda'] - merged_data['normalized_kda'].min()) / (merged_data['normalized_kda'].max() - merged_data['normalized_kda'].min())
```

```python
merged_data['Composite_Score'] = (
    0.5 * merged_data['normalized_winrate'] +
    0.3 * merged_data['normalized_popularity'] +
    0.2 * merged_data['normalized_kda']
)
```

```python
preferred_threshold = merged_data['Composite_Score'].quantile(0.75)
avoided_threshold = merged_data['Composite_Score'].quantile(0.25)
```

```python
def classify_champion(score):
    if score > preferred_threshold:
        return 'Preferred'
    elif score < avoided_threshold:
        return 'Avoided'
    else:
        return 'Neutral'
```

```python
merged_data['Preference'] = merged_data['Composite_Score'].apply(classify_champion)
```
